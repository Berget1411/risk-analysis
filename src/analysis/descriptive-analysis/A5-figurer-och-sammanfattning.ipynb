## Viktigaste slutsatserna

Det här är de mest användbara slutsatserna att bära med sig vidare till modelleringen:

- **Datakvalitet bekräftad** (A.1): Inga saknade värden, Duration i rimligt intervall (0, 1], inga negativa värden, konsistenta kategorinivåer mellan träning och test.
- **Skador är sällsynta** (A.2): ~98 % nollskador. Count-data med stark nollövervikt, passar Poisson eller negativ binomial.
- **Testdatan liknar träningsdatan** (A.2): Skadefrekvensen i 2025 ligger i samma storleksordning som 2021–2024. Modeller tränade på historisk data bör kunna appliceras rimligt.
- **Verksamhet spelar stor roll** (A.3): VVS och Takarbeten har klart högst skadefrekvens. Målare och Elektriker lägst.
- **Geografi spelar stor roll** (A.3): Storstad har klart högst frekvens, Småstad lägst.
- **Tidseffekter finns** (A.3): Svag uppåttrend 2022–2024, inte dramatisk men tillräcklig för att testa årseffekt i modellen.
- **Duration fungerar som exponering** (A.3): Skadefrekvensen per exponerat år är stabil över duration-grupper. Bör användas som offset, inte förklarande variabel.
- **Ekonomiska variabler korrelerade** (A.4): Omsättning och Försäkringsbelopp bär liknande information om företagsstorlek. Multikollinearitet måste hanteras vid variabelval.
- **Större företag = högre risk** (A.4): Monoton trend i decilanalys för Omsättning och Försäkringsbelopp. Självrisk har svagare samband.
- **Möjliga interaktioner** (A.3): VVS × Storstad har extra hög frekvens, men heatmappen ser till stor del additiv ut.

## Formulering för rapporten

Ni kan skriva något i stil med:

> Den deskriptiva analysen visar att skadefrekvensen i portföljen är låg totalt sett, men att det finns tydliga skillnader mellan verksamhetstyper, geografiska områden och år. VVS, takarbeten och storstadsmarknader uppvisar högre skadefrekvens än övriga grupper. Samtidigt är ekonomiska variabler som omsättning och försäkringsbelopp tydligt högerskevda och delvis korrelerade. Testdatan (2025) visar en liknande skadebild som träningsdatan. Detta motiverar en modellansats där skadeantal modelleras som count-data med hänsyn till exponering via Duration samt med riskklassificering efter verksamhet, geografi och tid.
